```python
df.drop(columns=['source','geolocation_source'],axis=1,inplace=True)
```

```python
df2=df.copy
```

```python
df3=df.copy()
Q1=df3.quantile(0)
Q3=df3.quantile(0.85)
IQR=Q3-Q1
print(IQR)
```

```python
data = df3[~((df3 < (Q1 - 1.5 * IQR)) |(df3 > (Q3 + 1.5 * IQR))).any(axis=1)]
print(data.shape)
```
